```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread

%matplotlib inline 
```

# Part a)

```python
def low_rank_approximation(X, rank):
    # YOUR CODE GOES HERE
    s,v,d = np.linalg.svd(X,full_matrices=False)
    v[rank:] = 0
    return s.dot(np.diag(v)).dot(d)
```

```python
face = imread("./data/face.jpg")
```

```python
face_rank5 = low_rank_approximation(face,5)
face_rank20 = low_rank_approximation(face,20)
face_rank100 = low_rank_approximation(face,100)
```

```python
plt.imshow(face_rank5,cmap='gray')
plt.title('5')
plt.savefig('face_rank5.jpg')
```

```python
plt.imshow(face_rank20,cmap='gray')
plt.title('20')
plt.savefig('face_rank20.jpg')
```

```python
plt.imshow(face_rank100,cmap='gray')
plt.title('100')
plt.savefig('face_rank100.jpg')
```

# Part c)

```python
def mse(img1, img2):
    # YOUR CODE GOES HERE 
    if img1.shape != img2.shape:
        print('error')
        return 0
    
    else:
        MSE = 0
        x,y = img1.shape
        for i in range(x):
            for j in range(y):
                MSE += (img1[i,j]-img2[i,j])*(img1[i,j]-img2[i,j])
        return MSE
```

```python
MSE_face = []
MSE_sky = []
for i in range(100):
    MSE_face.append(mse(face,low_rank_approximation(face,i+1)))
    MSE_sky.append(mse(sky,low_rank_approximation(sky,i+1)))
    print(i)
```

```python
plt.plot(range(1,101),MSE_face)
plt.plot(range(1,101),MSE_sky)
plt.legend(('face','sky'))
plt.savefig('mse.jpg')
```
